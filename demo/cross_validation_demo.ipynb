```python
import os
import sys
sys.path.append(os.getcwd() + '/../')
import plda
import numpy as np
```

```python
X = np.load('../cross_validation/preprocessed_cafe/X.npy')
Y = np.load('../cross_validation/preprocessed_cafe/Y.npy')
fnames = np.load('../cross_validation/preprocessed_cafe/fnames.npy')
```

```python
classifier = plda.Classifier(X, Y, fnames)
discriminator = plda.Discriminator(X, Y, fnames)
```

```python
results_classification, column_titles_classification = classifier.cross_validate(n=1, leave_out=True)
```

```python
results_discrimination, column_titles_discrimination = discriminator.cross_validate(n=3, leave_out=True)
```

```python
# Classification accuracy. Chance performance is .1429, and the model performs at .4111, in this case.
results = results_classification
(results[0,:,:,0] == results[0, :, :, 1]).sum() / np.prod(results.shape[:-1])
```

```python
# Discrimination accuracy (i.e. evaluating a pair of images as coming from the same category or not))
# Chance performance is .5, and the model performs at .8536, in this case.
results = results_discrimination
(results[0,:,:,0] == results[0, :, :, 1]).sum() / np.prod(results.shape[:-1])
```
